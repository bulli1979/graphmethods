```python
#general Imports
import networkx as nx
import math
import matplotlib.pyplot as plt
```

```python
#all functions for minoli count paths and edges and nodes
#count path Variables 
def getPathCount(G):
    nodes = G.nodes()
    pathCount = 0
    count = 1
    for nodeFrom in nodes:
        for nodeTo in nodes :
            if nodeTo != nodeFrom:                
                countForThis = getNumberOfPaths(nodeFrom,nodeTo,G)    
                pathCount += countForThis    
    return pathCount

#get number of Paths for node from and node To
def getNumberOfPaths(nodeFrom,nodeTo,G):
    foundVals = 0
    #remember the definition for proper paths minoli no node more than once save here still visited nodes
    visitedNodes = [nodeFrom]
    neighbors = list(G.neighbors(nodeFrom))
    
    while len(neighbors) > 0:
        newNeighbors = []
        for node in neighbors:
            if node == nodeTo:
                foundVals = foundVals + 1
            else:
                if node not in visitedNodes:
                    visitedNodes.append(node)
                    newNeighbors = newNeighbors + list(G.neighbors(node))
        neighbors = newNeighbors
    return foundVals
    
def minoli(G):
    nodes = nx.number_of_nodes(G)
    edges = nx.number_of_edges(G)
    #Minoli N * E / (N + E) * sum(oij)
    return (nodes * edges / (nodes + edges)  * getPathCount(G))
```

```python
#count all shortest paths together for wiener index
def wiener(G):
    indexCount = 0
    nodes = G.nodes()
    for nodeFrom in nodes:
        for nodeTo in nodes :
            if nodeTo != nodeFrom:                
                countForThis = getShortestPath(nodeFrom,nodeTo,G)    
                indexCount += countForThis    
    return indexCount

def getShortestPath(nodeFrom,nodeTo,G):
    pathRound = 0
    found = False
    #remember the definition for proper paths minoli no node more than once save here still visited nodes
    visitedNodes = [nodeFrom]
    neighbors = list(G.neighbors(nodeFrom))
    while len(neighbors) > 0 and found == False :
        newNeighbors = []
        pathRound+=1;
        for node in neighbors:
            if node == nodeTo:
                found = True
            else:
                if node not in visitedNodes:
                    visitedNodes.append(node)
                    newNeighbors = newNeighbors + list(G.neighbors(node))
        
        neighbors = newNeighbors
        
    if found:
        return pathRound
    return 0
```

```python
#longest shortest paths together for diameter index
def diameter(G):
    longestPath = 0
    nodes = G.nodes()
    startNode = None
    endNode = None
    for nodeFrom in nodes:
        for nodeTo in nodes :
            if nodeTo != nodeFrom:                
                countForThisNode = getLongestShortestPath(nodeFrom,nodeTo,G)
                if countForThisNode > longestPath:
                    longestPath = countForThisNode
                    startNode = nodeFrom
                    endNode = nodeTo
    print("StartNode: " + str(startNode) + " - EndNode: " + str(endNode))
    return longestPath

def getLongestShortestPath(nodeFrom,nodeTo,G):
    pathRound = 0
    found = False
    #remember the definition for proper paths minoli no node more than once save here still visited nodes
    visitedNodes = [nodeFrom]
    neighbors = list(G.neighbors(nodeFrom))
    while len(neighbors) > 0 and found == False :
        newNeighbors = []
        pathRound+=1;
        for node in neighbors:
            if node == nodeTo:
                found = True
            else:
                if node not in visitedNodes:
                    visitedNodes.append(node)
                    newNeighbors = newNeighbors + list(G.neighbors(node))
        neighbors = newNeighbors
    if found:
        return pathRound
    return 0
```
