```python
from pandas import DataFrame, read_csv
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
file = r'data/diagramData.csv'
df = pd.read_csv(file)
labels = df["LABEL"]
kk = df["kk"]
kkP1 = [kk[0],kk[1],kk[2]]
kkP2 = [kk[3],kk[4],kk[5],kk[6],kk[7]]
kkP3 = [kk[8],kk[9],kk[10]]
wiener = df["wiener"]
wienerP1 = [wiener[0],wiener[1],wiener[2]]
wienerP2 = [wiener[3],wiener[4],wiener[5],wiener[6],wiener[7]]
wienerP3 = [wiener[8],wiener[9],wiener[10]]
diameters = df["diameter"]
diametersP1 = [diameters[0],diameters[1],diameters[2]]
diametersP2 = [diameters[3],diameters[4],diameters[5],diameters[6],diameters[7]]
diametersP3 = [diameters[8],diameters[9],diameters[10]]
storyPoints = df["storyPoints"]
storyPointsP1 = [storyPoints[0],storyPoints[1],storyPoints[2]]
storyPointsP2 = [storyPoints[3],storyPoints[4],storyPoints[5],storyPoints[6],storyPoints[7]]
storyPointsP3 = [storyPoints[8],storyPoints[9],storyPoints[10]]
estimation = df["estimation"]
estimationP1 = [estimation[0],estimation[1],estimation[2]]
estimationP2 = [estimation[3],estimation[4],estimation[5],estimation[6],estimation[7]]
estimationP3 = [estimation[8],estimation[9],estimation[10]]
colors = ["red","red","red","green","green","green","green","green","blue","blue","blue"]
```

```python
fit = np.polyfit(kk, estimation,1)
fit_fn = np.poly1d(fit) 
plt.plot(kk,estimation, 'yo', kk, fit_fn(kk), '--k',label='Kombinatorische Komplexität - Aufwand')
plt.ylabel('Aufwand in Stunden')
plt.xlabel('Kombinatorisches Komplexitätsmass')
print (np.corrcoef(estimation,kk))
plt.show()
```

```python
fit = np.polyfit(wiener, estimation,1)
fit_fn = np.poly1d(fit) 
plt.plot(wiener,estimation, 'yo', wiener, fit_fn(wiener), '--k',label='Wiener Index - Aufwand')
plt.ylabel('Aufwand in Stunden')
plt.xlabel('Wiener Index')
print (np.corrcoef(estimation,wiener))
plt.show()
```

```python
fit = np.polyfit(diameters, estimation,1)
fit_fn = np.poly1d(fit) 
plt.plot(diameters,estimation, 'yo', diameters, fit_fn(diameters), '--k',label='Netzwerkdurchmesser - Aufwand')
plt.ylabel('Aufwand in Stunden')
plt.xlabel('Netzwerkdurchmesser')
print (np.corrcoef(estimation,diameters))
plt.show()
```
